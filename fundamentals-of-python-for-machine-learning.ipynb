```python
def isIran(text):
    if ('iran' in text.lower()):
        return True;
    else:
        return False;
    
isIran('Shiraz, Fars province, Iran')
```

# Practice No. 6
Create a function to give a ticket for high-speed driving based on the conditions below:
1. if speed is less than $60_\frac{KM}{h}$ *(speed < $60_\frac{KM}{h}$)*, "No ticket";
2. if speed is between $60_\frac{KM}{h}$ and $80_\frac{KM}{h}$ *($60_\frac{KM}{h}$ < speed < $80_\frac{KM}{h}$)*, "Small ticket";
3. if speed is more than $80_\frac{KM}{h}$ *(speed > $80_\frac{KM}{h}$)*, "Big tticket".

If it's the driver's birthday, give $5_\frac{KM}{h}$ more for any state.

```python
def ticket(speed,isBirthday):
    Birth = (5 if (isBirthday.lower()=='yes') else 0);
    
    if (speed <= (60 + Birth)):
        return 'No Ticket';
    elif (speed>(60 + Birth) and speed<=(80 + Birth)):
        return 'Small Ticket';
    else:
        return 'Big Ticket';

pd.DataFrame([
    ticket(65,'Yes'),
    ticket(65,'No'),
    
    ticket(85,'Yes'),
    ticket(85,'No'),
    
    ticket(90,'Yes'),
    ticket(90,'No')
],columns=["Ticket type"],index=[
    "ticket(65,'Yes')",
    "ticket(65,'No')",
    
    "ticket(85,'Yes')",
    "ticket(85,'No')",
    
    "ticket(90,'Yes')",
    "ticket(90,'No')"
])
```
